```python
df_transactions = pd.read_excel('Lumen_Data_V.xlsx', sheet_name = 'Donornet Transactions')
```

```python
df_contacts = pd.read_excel('Lumen_Data_V.xlsx', sheet_name = 'Donornet Contacts')
```

```python
df_contacts
```

```python
df_merged = df_transactions.merge(df_contacts, how = 'left', on = 'Email') 
```

```python
df_merged
```

```python
df_merged.columns
```

```python
df_merged.drop(['ID_x', 'Email', 'Item', 'Fund', 'Payment Method', 'Ref/Check#', 'Type',
       'Recurring Payment ID', 'ID_y', 'Firstname', 'Lastname', 'Phone',
       'Address1', 'Address2', 'Zip', 'Country',
       'Date of Birth', 'Do Not Solicit', 'Notes', 'Status', 'Created Date',
       'Modified Date'], axis=1, inplace=True)
```

```python
df_merged.to_csv('MapVis.csv',index=False)
```
